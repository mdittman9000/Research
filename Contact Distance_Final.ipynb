```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def Distance(a,E,V,step):

    V_MV = (V/1000)/1000
    
    d = V_MV / E
    d_um = d*10000
    print("Distance between contacts is " + str(d_um) + "Microns")
    
    #Shockley = Ohmic - (inputVoltage/Electric Field limit)
    a_cm = a/10000
    b_cm = a_cm + (V_MV/E)
    b = b_cm*10000
    print("Ohmic contact is " + str(b) +" Microns from center")
    
    # graph inner circle
    x = np.arange(-a,a,float(step))
    y = np.zeros(math.ceil(2*a*(1/float(step))))
    y_neg = np.zeros(math.ceil(2*a*(1/float(step))))
    
    for i in range(len(x)-1):
        y[i] = math.sqrt((a**2) - (x[i]**2))
        y_neg[i] = -math.sqrt((a**2) - (x[i]**2))
    
    plt.plot(x,y,color="Orange",label="Shockley")
    plt.plot(x,y_neg,color="Orange")
    
    #graph outer circle
    x2 = np.arange(-b,b,float(step))
    y2 = np.zeros(math.ceil(2*b*(1/float(step))))
    y2_neg = np.zeros(math.ceil(2*b*(1/float(step))))
    
    for i in range(len(x2)-1):
        y2[i] = math.sqrt((b**2) - (x2[i]**2))
        y2_neg[i] = -math.sqrt((b**2) - (x2[i]**2))
    
    plt.plot(x2,y2,color="Purple", label="Ohmic")
    plt.plot(x2,y2_neg,color="Purple")
    plt.legend()
```

```python
Distance(3,4,100,.01)
```
