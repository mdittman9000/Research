```python
import numpy as np
import matplotlib.pyplot as plt
import math
import ipywidgets as wg
```

```python
#this function really just asks; does this arrangement work?
#preferable we want it to tell us what arrangements work given the Electric field and Voltage limitations
```

```python
def ConcentricCircle(a,b,step):
    # x^2 + y^2 = r^2
    
    # graph inner circle
    x = np.arange(-a,a,float(step))
    y = np.zeros(math.ceil(2*a*(1/float(step))))
    y_neg = np.zeros(math.ceil(2*a*(1/float(step))))
    
    for i in range(len(x)-1):
        y[i] = math.sqrt((a**2) - (x[i]**2))
        y_neg[i] = -math.sqrt((a**2) - (x[i]**2))
    
    plt.plot(x,y,color="Orange",label="Shockley")
    plt.plot(x,y_neg,color="Orange")
    
    #graph outer circle
    x2 = np.arange(-b,b,float(step))
    y2 = np.zeros(math.ceil(2*b*(1/float(step))))
    y2_neg = np.zeros(math.ceil(2*b*(1/float(step))))
    
    for i in range(len(x2)-1):
        y2[i] = math.sqrt((b**2) - (x2[i]**2))
        y2_neg[i] = -math.sqrt((b**2) - (x2[i]**2))
    
    plt.plot(x2,y2,color="Red", label="Ohmic")
    plt.plot(x2,y2_neg,color="Red")
    plt.legend()
    
    #Return potential
    # V = E*d => V = -E*[si - sf] = -E*[a-b]
    # V = 100V
    # E = 4MV/cm = 4x10^8 [V/m]
    E1 = (100)/(b-a)
    print("E = "+str(E1) + "MV/cm")
    
```

```python
#input of a and b should be in microns
```

```python
ConcentricCircle(5,10,.01)
```
